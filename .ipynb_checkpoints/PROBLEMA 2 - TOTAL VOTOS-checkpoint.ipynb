```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import itertools
from time import time
from random import *
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import  r2_score, mean_absolute_error,mean_squared_error
from sklearn.linear_model import LinearRegression
import pickle
from sklearn.model_selection import train_test_split
```

```python
CAMINHO_DAS_BASES = 'C:/Users/felip/OneDrive/Documentos/tcc/HISTORICO_TCC2/BASES_DE_DADOS_FINALIZADAS/'

BASE_PREDICT_TOTAL_VOTOS_ML = pd.read_csv(CAMINHO_DAS_BASES+'BASE_PREDICT_TOTAL_VOTOS_ML_TRATADOS.txt',sep=';',encoding='ANSI')


TITULOS_X = ['ANO_ELEICAO', 'CODIGO_ESTADO', 'CODIGO_MUNICIPIO', 'NUMERO_PARTIDO',
       'ANALFABETO', 'ENSINO_FUNDAMENTAL_COMPLETO',
       'ENSINO_FUNDAMENTAL_INCOMPLETO', 'ENSINO_MEDIO_COMPLETO',
       'ENSINO_MEDIO_INCOMPLETO', 'LE_E_ESCREVE', 'SUPERIOR_COMPLETO',
       'SUPERIOR_INCOMPLETO', 'TOTAL_GASTO', 'SEXO', 'COD_GRAU_INSTRUCAO',
       'QTD_APTOS']


TITULO_Y = ['TOTAL_VOTOS']



X = BASE_PREDICT_TOTAL_VOTOS_ML[TITULOS_X]
y = BASE_PREDICT_TOTAL_VOTOS_ML[TITULO_Y]



X_Train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

 # 1 Âª FASE - RANDOM FOREST PARAMETROS

```python
# TESTES N_ESTIMADORES

def criar_objeto():
    results = {
            'R2_SCORE' : [],
            'MAE':[],
            'RMSE':[],
            'VALOR':[]
    }
    return results

def returnMetricas(results,y_predict,valor):
    
    MAE = mean_absolute_error(y_test,y_predict)
    RMSE = mean_squared_error(y_test,y_predict)
    R2 = r2_score(y_test,y_predict)
    
    results['R2_SCORE'].append(R2)
    results['VALOR'].append(valor)
    results['MAE'].append(MAE)
    results['RMSE'].append(RMSE)
    return results,MAE
        
def test_n_estimators(lista,results):
    num_test = len(lista)
    for i in lista:
        robo = RandomForestRegressor(n_estimators=i)
        robo.fit(X_Train,y_train)
        y_predict = robo.predict(X_test)
        
        results,MAE =returnMetricas(results,y_predict,i)
        
        
        num_test = num_test-1
        print("--ESTIMADOR-- N_ESTIMATORS ","TESTE PARAM : ", i, " MAE : ",MAE," FALTAM ",num_test)
    DATA_FRAME = pd.DataFrame(results).sort_values('VALOR')
    DATA_FRAME.to_csv("C:/Users/felip/OneDrive/Documentos/tcc/HISTORICO_TCC2/RESULTADOS_PARAMETROS/RANDOM_FOREST/PROBL_2/N_ESTIMATORS_PARAM_2.txt",sep=";",encoding="ANSI",index_label=False)
    return DATA_FRAME

def test_max_depth(lista,results):
    num_test = len(lista)
    for i in lista:
        robo = RandomForestRegressor(max_depth=i)
        robo.fit(X_Train,y_train)
        y_predict = robo.predict(X_test)
        
        results,MAE =returnMetricas(results,y_predict,i)
        
        
        num_test = num_test-1
        print("--ESTIMADOR-- MAX_DEPTH_PARAM ","TESTE PARAM : ", i, " MAE : ",MAE," FALTAM ",num_test)
    DATA_FRAME = pd.DataFrame(results).sort_values('VALOR')
    DATA_FRAME.to_csv("C:/Users/felip/OneDrive/Documentos/tcc/HISTORICO_TCC2/RESULTADOS_PARAMETROS/RANDOM_FOREST/PROBL_2/MAX_DEPTH_PARAM_2.txt",sep=";",encoding="ANSI",index_label=False)
    return DATA_FRAME

def test_min_samples_split(lista,results):
    num_test = len(lista)
    for i in lista:
        robo = RandomForestRegressor(min_samples_split=i)
        robo.fit(X_Train,y_train)
        y_predict = robo.predict(X_test)
        
        results,MAE =returnMetricas(results,y_predict,i)
        
        
        num_test = num_test-1
        print("--ESTIMADOR-- MIN_SAMPLES_SPLIT_PARAM ","TESTE PARAM : ", i, " MAE : ",MAE," FALTAM ",num_test)
    DATA_FRAME = pd.DataFrame(results).sort_values('VALOR')
    DATA_FRAME.to_csv("C:/Users/felip/OneDrive/Documentos/tcc/HISTORICO_TCC2/RESULTADOS_PARAMETROS/RANDOM_FOREST/PROBL_2/MIN_SAMPLES_SPLIT_PARAM_2.txt",sep=";",encoding="ANSI",index_label=False)
    return DATA_FRAME

def test_min_samples_leafh(lista,results):
    num_test = len(lista)
    for i in lista:
        robo = RandomForestRegressor(min_samples_leaf=i)
        robo.fit(X_Train,y_train)
        y_predict = robo.predict(X_test)
        
        results,MAE =returnMetricas(results,y_predict,i)
        
        
        num_test = num_test-1
        print("--ESTIMADOR-- MIN_SAMPLES_LEAFTH_PARAM ","TESTE PARAM : ", i, " MAE : ",MAE," FALTAM ",num_test)
        
    DATA_FRAME = pd.DataFrame(results).sort_values('VALOR')
    DATA_FRAME.to_csv("C:/Users/felip/OneDrive/Documentos/tcc/HISTORICO_TCC2/RESULTADOS_PARAMETROS/RANDOM_FOREST/PROBL_2/MIN_SAMPLES_LEAFTH_PARAM_2.txt",sep=";",encoding="ANSI",index_label=False)
    return DATA_FRAME

def test_max_features(lista,results):
    num_test = len(lista)
    for i in lista:
        robo = RandomForestRegressor(max_features=i)
        robo.fit(X_Train,y_train)
        y_predict = robo.predict(X_test)
        
        results,MAE =returnMetricas(results,y_predict,i)
        
        
        num_test = num_test-1
        print("--ESTIMADOR-- MAX_FEATURES_PARAM ","TESTE PARAM : ", i, " MAE : ",MAE," FALTAM ",num_test)
    DATA_FRAME = pd.DataFrame(results).sort_values('VALOR')
    DATA_FRAME.to_csv("C:/Users/felip/OneDrive/Documentos/tcc/HISTORICO_TCC2/RESULTADOS_PARAMETROS/RANDOM_FOREST/PROBL_2/MAX_FEATURES_PARAM_2.txt",sep=";",encoding="ANSI",index_label=False)
    return DATA_FRAME

def start_test_random():
    print("Iniciando testes N_estimators : ")
    test_n_estimators([x for x in range(2,600,25)],criar_objeto())
    print("Iniciando testes max depth : ")
    test_max_depth([x for x in range(2,150,2)],criar_objeto())
    print("Iniciando testes split : ")
    test_min_samples_split([x for x in range(2,200,2)],criar_objeto())
    print("Iniciando testes leaf : ")
    test_min_samples_leafh([x for x in range(2,200,1)],criar_objeto())
    print("Iniciando testes features : ")
    test_max_features(['auto','sqrt','log2',None],criar_objeto())
    print("FINALIZADO , TESTADO E SALVO TODOS OS TESTES, PODE COMEMORAR !!! ")
    
start_test_random()
```
