```python
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
imp_moda = SimpleImputer(missing_values=np.nan, strategy='most_frequent')

TOTAL_VOTOS_E_GASTOS = pd.DataFrame(imp_mean.fit_transform(BASE_CLASSIFICADOR_ELEITO_ML[['TOTAL_VOTOS','TOTAL_GASTO']]))
SEXO_COD_INSTRUCAO = pd.DataFrame(imp_moda.fit_transform(BASE_CLASSIFICADOR_ELEITO_ML[['SEXO','COD_GRAU_INSTRUCAO']]))

TOTAL_VOTOS_E_GASTOS.columns = ['TOTAL_VOTOS','TOTAL_GASTO']
SEXO_COD_INSTRUCAO.columns   = ['SEXO','COD_GRAU_INSTRUCAO']
```

```python
BASE_CLASSIFICADOR_ELEITO_ML_TRATADOS = BASE_CLASSIFICADOR_ELEITO_ML

del BASE_CLASSIFICADOR_ELEITO_ML_TRATADOS['TOTAL_VOTOS'],BASE_CLASSIFICADOR_ELEITO_ML_TRATADOS['TOTAL_GASTO'],BASE_CLASSIFICADOR_ELEITO_ML_TRATADOS['SEXO'],BASE_CLASSIFICADOR_ELEITO_ML_TRATADOS['COD_GRAU_INSTRUCAO']

BASE_CLASSIFICADOR_ELEITO_ML_TRATADOS['TOTAL_VOTOS'] = TOTAL_VOTOS_E_GASTOS.TOTAL_VOTOS.values
BASE_CLASSIFICADOR_ELEITO_ML_TRATADOS['TOTAL_GASTO'] = TOTAL_VOTOS_E_GASTOS.TOTAL_GASTO.values
BASE_CLASSIFICADOR_ELEITO_ML_TRATADOS['SEXO'] = SEXO_COD_INSTRUCAO.SEXO.values
BASE_CLASSIFICADOR_ELEITO_ML_TRATADOS['COD_GRAU_INSTRUCAO'] = SEXO_COD_INSTRUCAO.COD_GRAU_INSTRUCAO.values
```

```python
#VERIFICANDO VALORES NULOS
plt.figure(figsize=(20,8))
sns.heatmap(BASE_CLASSIFICADOR_ELEITO_ML_TRATADOS.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```
